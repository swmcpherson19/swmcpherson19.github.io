```python
# M5:Trump's Tweets Assignment - Scott McPherson
# importing libraries and files into data
import matplotlib.pyplot as plt
f_in = open ('C:/Users/Owner/Documents/BUAD 502C/Module5/words.csv','r')
data = f_in.readlines()
f_in.close()

""" The list below contains words that you are to exclude form your analysis """
stopWords = ['to', 'the', 'a', 'to', 'is', 'and', 'in', 'of', 'that', 'it', 'be','at', 'this', 'are', 'be', 'for', 'will', 'with', 'at', 'have','on','&amp', 'by']

# data wrangling section
for i in range(len(data)):
    data[i]=data[i].strip()
data = [x for x in data if x not in stopWords]

# dictionary setup -- tried to do dictionary comp., but the kernel kept crashing
tweetDict={}
for i in data:
    if i in tweetDict:
        tweetDict[i] += 1
    else:
        tweetDict[i] = 1

# sorting the dictionary by values to get the highest frequency at the beginning
sortedtweetDict={k: v for k,v in sorted(tweetDict.items(), key=lambda item: item[1], reverse=True)}

# removing the top ten of each the keys and values
keylist = list(sortedtweetDict.keys())
valuelist = list(sortedtweetDict.values())
x = keylist[:10]
xs = range(len(x))  # this is for use on the xticks later -- requires an integer to set up the number of ticks
y = valuelist[:10]

#set up subplots, axes, and figures
fig,ax = plt.subplots()
ax.bar(xs,y, align='center', color='r')
ax.set_xticks(xs)
ax.set_xticklabels(x, rotation=45)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_label_text("Words in Trump's Tweets", fontsize=14)
ax.yaxis.set_label_text("Number of Occurrences", fontsize=14)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.tick_params(axis = 'x', which = 'both', direction = 'in', width = 2, color = 'black')
fig.suptitle("Top 10 Most Frequently Tweeted Words by President Trump")
fig.set_size_inches(16,8)
fig.savefig('tweets.jpg', bbox_inches='tight')

# displaying graph
plt.show()
```
